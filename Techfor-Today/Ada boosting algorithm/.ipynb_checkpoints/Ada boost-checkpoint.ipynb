```python
# importing the required modules
from sklearn import datasets
import pandas as pd
import numpy as np

# loading the iris dataset
dataset = datasets.load_iris()

# converting the data to DataFrame
data = pd.DataFrame(data= np.c_[dataset['data'], dataset['target']],
                     columns= dataset['feature_names'] + ['target'])

# printing the few rows
data.head()
```

```python
# plotting the scatter plot based on coloring
data.plot.scatter(x="sepal length (cm)", y="sepal width (cm)", c='target', s=100, figsize=(10, 6))
```

```python
# plotting the box plot
data.drop('target', axis=1).plot.box(figsize=(10, 6))
```

```python
# splitting the dataset into input and output
Input = data.drop('target', axis=1)
Output =data['target']
```

```python
# importing the module
from sklearn.model_selection import train_test_split

# splitting into testing and training parts
X_train, X_test, y_train, y_test = train_test_split(Input, Output, test_size=0.25)
```

```python
# importing the classifier
from sklearn.ensemble import AdaBoostClassifier

# initailizing the ada boost classifier with 2 stump trees
Ada_clf = AdaBoostClassifier(n_estimators=1)
```

```python
# training the classifier
AdaBoost_2stump = Ada_clf.fit(X_train, y_train)
```

```python
# making predictions
AdaBoost_pred = AdaBoost_2stump.predict(X_test)
```

```python
# importing modules
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# confusion matrix plotting
cm = confusion_matrix(y_test,AdaBoost_pred, labels=Ada_clf.classes_)

# labelling
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=Ada_clf.classes_)
disp.plot()
plt.show()
```

```python
# importing the module
from sklearn.metrics import accuracy_score

# printing
print("The accuracy is:  ", accuracy_score(y_test, AdaBoost_pred))
```

```python
# initailizing the ada boost classifier with 10 stump trees
Ada_clf = AdaBoostClassifier(n_estimators=10)

# training the classifier
AdaBoost_10stump = Ada_clf.fit(X_train, y_train)
```

```python
# making predictions
AdaBoost_pred = AdaBoost_10stump.predict(X_test)
```

```python
# confusion matrix plotting
cm = confusion_matrix(y_test,AdaBoost_pred, labels=Ada_clf.classes_)

# labelling
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=Ada_clf.classes_)
disp.plot()
plt.show()
```

```python
# printing
print("The accuracy is:  ", accuracy_score(y_test, AdaBoost_pred))
```

```python
# importing dataset 
data = pd.read_csv('house.csv')

# heading of the dataset
data.head()
```

```python
# removing the null values
data.dropna(axis=0, inplace = True)
```

```python
# input and output variables
Input = data.drop('price', axis=1)
Output = data.price
```

```python
# importing the module
from sklearn.model_selection import train_test_split

# splitting into testing and training parts
X_train, X_test, y_train, y_test = train_test_split(Input, Output, test_size=0.25)
```

```python
# importing ada boost regressor
from sklearn.ensemble import AdaBoostRegressor

# Create adaboot with 20 stump trees
Ada_regressor = AdaBoostRegressor(n_estimators=20)
```

```python
# training the ada boost regressor
AdaBoost_R = Ada_regressor.fit(X_train, y_train)
```

```python
#Predict price 
AdaBoost_pred = AdaBoost_R.predict(X_test)
```

```python
#  fitting the size of the plot
plt.figure(figsize=(12, 8))

# plotting the graph for the actual values
plt.plot([i for i in range(len(y_test))],y_test, label="actual values")

# plotting the graph for predictions
plt.plot([i for i in range(len(y_test))],AdaBoost_pred, label="Predicted values", c='m')

plt.legend()
plt.show()
```

```python
# importing all the necessary modules
from numpy import mean
from numpy import std
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from matplotlib import pyplot


# get a list of models to evaluate
def get_models():
	models = dict()
	# define number of trees to consider
	n_trees = [10, 50, 100, 500, 1000]
	for n in n_trees:
		models[str(n)] = AdaBoostClassifier(n_estimators=n)
	return models

# evaluate a given model using cross-validation
def evaluate_model(model, X, y):
	# define the evaluation procedure
	cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
	# evaluate the model and collect the results
	scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
	return scores

# define dataset
X  = Input
y = Output
# get the models to evaluate
models = get_models()
# evaluate the models and store results
results, names = list(), list()
for name, model in models.items():
	# evaluate the model
	scores = evaluate_model(model, X, y)
	# store the results
	results.append(scores)
	names.append(name)
	# summarize the performance along the way
	print('>%s %.3f (%.3f)' % (name, mean(scores), std(scores)))
# plot model performance for comparison
pyplot.boxplot(results, labels=names, showmeans=True)
pyplot.show()
```

```python
# defining the classifier
model = AdaBoostClassifier()

# creating a dic for the grid
grid = dict()

# estimator till 500
grid['n_estimators'] = [10, 50, 100, 200, 500]

# defining learning rate
grid['learning_rate'] = [0.0001, 0.01, 0.1, 1.0, 1.1, 1.2]


# defining the CV
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3)

# initializing the grid search
grid_search = GridSearchCV(estimator=model, param_grid=grid, n_jobs=-1, cv=cv, scoring='accuracy')


# training the model on grid search
grid_result = grid_search.fit(Input, Output)

# finding the best results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```
