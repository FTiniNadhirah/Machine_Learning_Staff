```python
# importing the required modules
import numpy as np
import matplotlib.pyplot as plt

# creating random x and y variables 
x = np.random.normal(size=(1, 1000))
y = np.random.normal(size=(1, 1000))

# this line will produce 
plt.hexbin(x, y, gridsize=25, cmap="twilight_shifted")
```

```python
# building hexagons 
plt.hexbin(x,y,gridsize=15, mincnt=1, edgecolors="none", cmap="twilight_shifted")

# visualizing the datapoints
plt.scatter(x,y, s=2, c="white")
plt.show()
```

```python
# importing the required module
import seaborn as sns

# setting the theme style
sns.set_theme(style="whitegrid")

# creating random numbers
rs = np.random.RandomState(11)
x = rs.gamma(2, size=1000)
y = -.5 * x + rs.normal(size=1000)


# plotting the plots using seaborn
sns.jointplot(x=x, y=y, kind="hex", color="#9560FD")
```

```python
# importing the module
import geojson

# opening the json file and loading
with open('Bishkek.json') as f:
    file = geojson.load(f)
```

```python
# importing the module'
import geopandas as gpd

# converting the json to GeoDataFrame
gd_file = gpd.GeoDataFrame(file.features)


# adding few preprocessing steps
gd_file['id'] = [gd_file['properties'][i]['id'] for i in range(gd_file.shape[0])]
gd_file.drop(['type', 'properties'], axis=1, inplace=True)
gd_file.head()
```

```python
# plotting the hexagons
gd_file.plot()
```

```python
# importing folium
import folium

# creating the map
map = folium.Map(location=[42.85000, 74.59000], zoom_start=11, tiles='CartoDB positron')

# adding the json file to the map
folium.GeoJson(file).add_to(map)

# showing the map
map
```
